# **One Hot Encoding Of Data**

```python
X_one_encode = pd.get_dummies(X, columns=['model' , 'transmission' , 'fuelType'] , drop_first=True)
X_one_encode
```

```python
X_one_encode = X_one_encode.astype(int)
X_one_encode
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
numeric_cols = ['year', 'mileage', 'tax', 'mpg', 'engineSize']
X_one_encode[numeric_cols] = scaler.fit_transform(X_one_encode[numeric_cols])
X_one_encode
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import Ridge , Lasso

X_train , X_test , Y_train , Y_test = train_test_split(X_one_encode , Y , test_size=0.2 , random_state=42)
```

```python
ridge = Ridge(alpha = 1.0)  # λ = 1.0
ridge.fit(X_train , Y_train)

y_pred_ridge = ridge.predict(X_test)
y_pred_ridge
```

```python
lasso = Lasso(alpha = 0.05)
lasso.fit(X_train , Y_train)

y_pred_lasso = lasso.predict(X_test)
y_pred_lasso
```

# **For Ridge**

```python
from sklearn.metrics import mean_squared_error , r2_score

r2 = r2_score(Y_test , y_pred_ridge);
print("R2 Score : " , r2)

n = X_test.shape[0]
p = X_test.shape[1]
adjusted_r2 = 1 - ((1 - r2) * (n - 1))/(n - p - 1)
print("Adjusted R2 Score : " , adjusted_r2)

mse = mean_squared_error(Y_test , y_pred_ridge)
print("Mean Squared Error : " , mse)
```

# **For Lasso**

```python
r2 = r2_score(Y_test , y_pred_lasso);
print("R2 Score : " , r2)

n = X_test.shape[0]
p = X_test.shape[1]
adjusted_r2 = 1 - ((1 - r2) * (n - 1))/(n - p - 1)
print("Adjusted R2 Score : " , adjusted_r2)

mse = mean_squared_error(Y_test ,y_pred_lasso)
print("Mean Squared Error : " , mse)
```

# **Label Enoding Of Data**

```python
from sklearn.preprocessing import LabelEncoder

label = LabelEncoder()
cols = ['model' , 'transmission' , 'fuelType']
Xlabel = X.copy()

for col in cols:
  Xlabel[col] = label.fit_transform(Xlabel[col])
Xlabel
```

```python
Xlabel = Xlabel.astype(int)
Xlabel
```

```python
Cols = ['model', 'year', 'transmission', 'mileage', 'fuelType', 'tax', 'mpg',
       'engineSize']
Xlabel[Cols] = scaler.fit_transform(Xlabel[Cols])
Xlabel
```

```python
X_test , X_train , Y_test , Y_train = train_test_split(Xlabel , Y , test_size=0.33 , random_state=42)
```

```python
ridge = Ridge(alpha = 1.0)
ridge.fit(X_train , Y_train)
y_pred_ridge = ridge.predict(X_test)
y_pred_ridge
```

```python
lasso = Lasso(alpha=0.05)
lasso.fit(X_train , Y_train)
y_pred_lasso = lasso.predict(X_test)
y_pred_lasso
```

# **For Ridge**

```python
r2 = r2_score(Y_test , y_pred_ridge);
print("R2 Score : " , r2)

n = X_test.shape[0]
p = X_test.shape[1]

adjusted_r2 = 1 - ((1 - r2) * (n - 1))/(n - p - 1)
print("Adjusted R2: " , adjusted_r2)

mse = mean_squared_error(Y_test ,y_pred_ridge)
print("Mean Squared Error : " , mse)
```

# **For Lasso**

```python
r2 = r2_score(Y_test , y_pred_lasso);
print("R2 Score : " , r2)

n = X_test.shape[0]
p = X_test.shape[1]

adjusted_r2 = 1 - ((1 - r2) * (n - 1))/(n - p - 1)
print("Adjusted R2: " , adjusted_r2)

mse = mean_squared_error(Y_test ,y_pred_lasso)
print("Mean Squared Error : " , mse)
```

# **Feature Selection using Ridge and Lasso**

```python
ridge_coeff = pd.Series(
    ridge.coef_, index=X.columns
).sort_values(key=abs, ascending=False)

ridge_coeff.head(10)
#Ridge shrinks coefficients but keeps all features.
```

```python
lasso_coeff = pd.Series(
    lasso.coef_, index=X.columns
)

selected_features = lasso_coeff[lasso_coeff != 0].sort_values(key=abs, ascending=False)

selected_features
```

```python
print("Total features:", X.shape[1])
print("Features selected by Lasso:", selected_features.shape[0])
```

**A positive coefficient means the feature increases the predicted car price, while a negative coefficient means it decreases the price, assuming all other features remain constant. When features are scaled, the sign indicates the effect relative to the feature’s average value.**
