# **Questions**

1. Which team won the most matches?

```python
data = df['match_winner'].value_counts()
data
```

```python
sns.barplot(y=data.index , x=data.values , palette='viridis')
plt.title("Most Match Win by Team")
```

2. Toss Decision Trends

```python
sns.countplot(x =df['toss_decision'] , palette='viridis')
plt.title("Toss Decision")
```

```python
df['toss_winner'].value_counts()
```

# **Key Player Performance**

1. Most "Player of the Match" Awards ?

```python
cnt = df['player_of_the_match'].value_counts().head()
cnt
```

```python
sns.barplot(y=cnt.index , x=cnt.values , palette='viridis')
plt.title("Most Player of the Match Awards")
```

2. Top Scorers

```python
cnt = df['top_scorer'].value_counts().head()
cnt
```

```python
sns.barplot(y=cnt.index , x=cnt.values , palette='viridis')
plt.title('Top Scorers')
```

```python
cnt = df.groupby('top_scorer')['highscore'].sum().sort_values(ascending=False).head()
cnt
```

```python
cnt.plot(kind='bar',color='red')
plt.title('Top Scorers')
```

3. Best Bowling Figures

```python
df['highest_wickets'] = df['best_bowling_figure'].apply(lambda x : x.split('--')[0])
df['highest_wickets'] = df['highest_wickets'].astype('int')
top_bowlers = df.groupby('best_bowling')['highest_wickets'].sum().sort_values(ascending=False).head()
top_bowlers.plot(kind='bar',color='green')
plt.title('Best Bowling Figures')
```
