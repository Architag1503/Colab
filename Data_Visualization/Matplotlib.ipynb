# **Object Oriented Way**

Matplotlib allows the apsect ratio , DPI and the figure size to be specified when the figure object is created. You can use the figsize and dpi keyword arguments.

figsize is a tuple of the width and height of the figure in inches dpi is the dots-per-inch (pixel per inch)

```python
fig = plt.figure(figsize = (12 , 8), dpi=150)
axis1 = fig.add_axes([0.1 , 0.1 , 1 , 1])
#1 paremeter -> how you want to shift from left
#2 parametr -> how you want to shift from bottom
#3 parameter -> percentage of width
#4 parameter -> percentage of height
axis1.plot(x , y)

axis2 = fig.add_axes([0.3 , 0.4 , 0.3 , 0.3])
axis2.plot(x , y)
```

In above image we can see that axis2 overlap the axis1 and axis2 shifted 30% from left and 40% from bottom. By this we can say that using object oriented way we have full control on canva. We can addjust the size of each canva as we want (shown below)

```python
fig = plt.figure(figsize = (8 , 15) , dpi=500)
axis1 = fig.add_axes([0.8 , 0.6 , 0.3 , 0.8])
axis1.plot(x , y)
axis2 = fig.add_axes([0.3 , 0.4 , 0.3 , 0.3])
axis2.plot(x , y)
```
