Using seaborn to plot categorical data. There are few main plot types for this:



*  boxplot
*  violinplot


*  stripplot
*  swarmplot


*  barplot
*  countplot







```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
df = sns.load_dataset('tips')
df
```

# **CountPlot**

```python
sns.countplot(x='sex', data=df , hue='smoker')
```

```python
sns.barplot(x=df['sex'] , y=df['total_bill'])
#x-axis represent categorical data
#y-axis represent cantinous values
```

We can change the estimator object to our own function, that converts a vector to a scalar

Mean (default): np.mean<br>
Median: np.median<br>
Sum: np.sum<br>
Count: len (to count the number the of observation)<br>
standard deviation: np.std<br>
Minimum: np.min<br>
maximum: np.max




```python
sns.barplot(x=df['sex'] , y=df['tip'] , estimator=np.sum)
```

# **BoxPlot**

```python
sns.boxplot(x = 'tip' , y = 'day' , data = df , palette="rainbow")
#x-axis represent the continous data
#y-axis reprsent the categorical data
```

boxplot and violinplot are used to shown the distribution of categorical data. A box plot (or box-and-whisker plot) shows the distribution of quantitative data in a way that facillitates comparison between variables or across levels of a categorical variable. The box shows the quartiles of the dataset while the whiskers extend to show the rest of the distribution , except for points that are determined to be "outliers" using a method that is a function of the inter-quartile range

# **ViolinPlot**

```python
sns.violinplot(x = 'day' , y = 'total_bill' , data = df , palette="rainbow")
#categorical data on x-axis
#continous data on y-axis
```

# **StripPlot & SwarmPlot**

The stripplot will draw a scatterplot where one variable is categorical. A strip plot can drawn on its own , but it is also good compliment to a box or violin plotin cases where you want to show all observations along with some representation of underlying distribution.

The swarmplot is similar to stripplot(), but the points are adjusted (only along the categorical axis) so that they dont overlap. This gives a better representation of the distribution of the values , although it does not scale as well to large numbers of observation(both in term of the ability to show all the points and in terms of the computation needed to arrange them)

```python
df = sns.stripplot(x = 'day' , y = 'total_bill' , data = df)
df
```

```python

df = sns.load_dataset("tips")   # df must be DataFrame

sns.swarmplot(x='day', y='total_bill', data=df)
```

```python
sns.violinplot(x = 'tip' , y = 'day' , data = df)
```

```python
sns.violinplot(x = 'tip' , y = 'day' , data = df)
sns.swarmplot(x = 'tip' , y = 'day' , data = df)
```
