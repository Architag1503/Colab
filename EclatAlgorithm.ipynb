# Scratch **Version**

```python
def convert_to_vertical(transactions):
    vertical = {}

    for tid, items in enumerate(transactions, start=1):
        for item in items:
            if item not in vertical:
                vertical[item] = set()
            vertical[item].add(tid)

    return vertical
```

```python
def eclat(prefix, items, min_support, result):
    """
    prefix: current itemset
    items: list of (item, tidset)
    min_support: minimum support threshold
    result: store final frequent itemsets
    """

    while items:
        item, tidset = items.pop(0)

        new_prefix = prefix + [item]
        support = len(tidset)

        # Check support
        if support >= min_support:
            result.append((new_prefix, support))

            # Generate new combinations
            new_items = []
            for other_item, other_tidset in items:
                intersection = tidset & other_tidset

                if len(intersection) >= min_support:
                    new_items.append((other_item, intersection))

            # Recursive DFS call
            eclat(new_prefix, new_items, min_support, result)
```

```python
# -------------------------------
# Main Function
# -------------------------------

def run_eclat():
    # Transactions (given dataset)
    transactions = [
        ['Bread', 'Butter', 'Milk'],
        ['Bread', 'Coke'],
        ['Milk', 'Coke', 'Jam'],
        ['Bread', 'Butter', 'Coke'],
        ['Bread', 'Milk', 'Jam'],
        ['Butter', 'Milk', 'Coke'],
        ['Bread', 'Butter', 'Milk', 'Jam'],
        ['Milk', 'Coke']
    ]

    min_support = 2

    # Step 1: Convert to vertical format
    vertical_data = convert_to_vertical(transactions)

    print("🔷 Vertical Format (Item → TID List):")
    for item in vertical_data:
        print(f"{item} → {vertical_data[item]}")
    print()

    # Step 2: Prepare items list
    items = [(item, tidset) for item, tidset in vertical_data.items()]

    # Sort items (optional but good practice)
    items.sort(key=lambda x: len(x[1]))

    result = []

    # Step 3: Run ECLAT
    eclat([], items, min_support, result)

    # Step 4: Print Results
    print("✅ Frequent Itemsets:\n")
    for itemset, support in result:
        print(f"{itemset} : {support}")
```

```python
# -------------------------------
# Execute
# -------------------------------

if __name__ == "__main__":
    run_eclat()
```

```python
import pandas as pd
from pyECLAT import ECLAT

def run_eclat_with_library():
    # -------------------------------
    # Step 1: Define Transactions
    # -------------------------------
    transactions = [
        ['Bread', 'Butter', 'Milk'],
        ['Bread', 'Coke'],
        ['Milk', 'Coke', 'Jam'],
        ['Bread', 'Butter', 'Coke'],
        ['Bread', 'Milk', 'Jam'],
        ['Butter', 'Milk', 'Coke'],
        ['Bread', 'Butter', 'Milk', 'Jam'],
        ['Milk', 'Coke']
    ]

    # -------------------------------
    # Step 2: Convert to DataFrame
    # -------------------------------
    df = pd.DataFrame(transactions)

    print("🔷 Input DataFrame:\n")
    print(df)
    print("\n" + "-"*50)

    # -------------------------------
    # Step 3: Apply ECLAT
    # -------------------------------
    eclat = ECLAT(data=df)

    # Convert min_support count → percentage
    min_support = 3 / len(transactions)

    indices, support = eclat.fit(
        min_support=min_support,
        min_combination=1,
        max_combination=3,
        separator=','
    )

    # -------------------------------
    # Step 4: Print Results
    # -------------------------------
    print("\n✅ Frequent Itemsets:\n")

    for itemset, sup in support.items():
        print(f"{itemset} : {round(sup * len(transactions))} (support count), "
              f"{round(sup, 3)} (support)")
```

```python
# -----------------------------------
# Run Program
# -----------------------------------
if __name__ == "__main__":
    run_eclat_with_library()
```
