# **Logistic Regression , KNN , Naive Bays , Decision Tree , SVM & Cross-Validation**

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

# **Logistic Regression**

```python
df = sns.load_dataset('titanic')
df
```

```python
df.shape
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
df.drop(["deck" , "embark_town" , "alive" , "class" , "who" , "adult_male"] , axis=1 , inplace=True)
```

```python
df.columns
```

```python
df["age"].fillna(df["age"].mean() , inplace=True)
```

```python
df.dropna(subset=["embarked"] , inplace=True)
```

```python
df.info()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()

df["sex"] = le.fit_transform(df["sex"])
df["embarked"] = le.fit_transform(df["embarked"]);
```

```python
df = df.astype(int)
```

```python
df
```

```python
X = df.drop(["survived"] , axis = 1)
Y = df["survived"]
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X = scaler.fit_transform(X)
```

```python
X
```

```python
from sklearn.model_selection import train_test_split
X_train , X_test , Y_train , Y_test = train_test_split(X , Y , test_size = 0.2 , random_state = 42)
```

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression()
model.fit(X_train , Y_train)
```

```python
y_pred = model.predict(X_test)
y_pred
```

```python
Y_test
```

```python
from sklearn.metrics import accuracy_score , confusion_matrix, classification_report
accuracy_score(Y_test , y_pred)
```

```python
confusion_matrix(Y_test , y_pred)
```

```python
print(classification_report(y_pred , Y_test))
```

# **K-Nearest Neighbour (KNN)**

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_train_scaled
```

```python
X_test_scaled = scaler.fit_transform(X_test)
X_test_scaled
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn_model = KNeighborsClassifier(n_neighbors=5)
knn_model.fit(X_train_scaled , Y_train)
```

```python
Y_knn_pred = knn_model.predict(X_test_scaled)
```

```python
from sklearn.metrics import accuracy_score , confusion_matrix , classification_report
accuracy_score(Y_test , Y_knn_pred)
```

```python
confusion_matrix(Y_knn_pred , Y_test)
```

```python
print(classification_report(Y_knn_pred , Y_test))
```

# **Naive Bays**

```python
y_pred_nb = nb_model.predict(X_test)
y_pred_nb
```

```python
from sklearn.metrics import accuracy_score , confusion_matrix , classification_report
accuracy_score(Y_test , y_pred_nb)
```

```python
from sklearn.metrics import accuracy_score , confusion_matrix , classification_report
accuracy_score(Y_test , y_pred_bn)
```

# **Decision Tree**

```python
from sklearn.tree import DecisionTreeClassifier
tree_model = DecisionTreeClassifier(random_state=42)
tree_model.fit(X_train_scaled , Y_train)
```

```python
y_pred_tree = tree_model.predict(X_test_scaled)
y_pred_tree
```
