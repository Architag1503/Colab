```python
X = df.drop(['species'] , axis = 1)
Y = df['species']
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X = scaler.fit_transform(X)
X
```

```python
from sklearn.model_selection import train_test_split

X_train , X_test , Y_train , Y_test = train_test_split(X , Y , test_size = 0.33 , random_state = 42)
```

```python
from sklearn.neighbors import KNeighborsClassifier
model_knn = KNeighborsClassifier(n_neighbors = 13)
model_knn.fit(X_train , Y_train)
```

```python
model_knn.score(X_test , Y_test)
```

```python
Y_pred = model_knn.predict(X_test)
Y_pred
```

```python
from sklearn.svm import SVC
```

```python
model_svm = SVC( gamma = 'auto')
model_svm.fit(X_train , Y_train)
```

```python
model_svm.score(X_test , Y_test)
```

```python
classifier = GridSearchCV((model_svm) , {
    'C' : [1,10,20,30],
    'kernel' : ['linear' , 'rbf']
} , cv = 5 , return_train_score = False)
```

```python
classifier.fit(X, Y)
```

# **2. Randomized Search CV**

```python
from sklearn.model_selection import RandomizedSearchCV
classifier3 = RandomizedSearchCV((model_knn) , {
    'n_neighbors' : range(1 , 21),
    'weights' : ['uniform' , 'distance'],
    'algorithm' : ['auto' , 'ball_tree' , 'kd_tree' , 'brute']
} , n_iter = 10 , cv = 5 , return_train_score = False)
```

```python
classifier3.fit(X , Y)
```

```python
classifier3.cv_results_
```

```python
results = pd.DataFrame(classifier3.cv_results_)
results = results.sort_values(by='rank_test_score')
results
```

**In above code we saw K-fold Cross Validation as we gave numeric value to "cv" parameter**

# **StratifiedKFold CV**

```python
from sklearn.model_selection import StratifiedKFold , RandomizedSearchCV
sf = StratifiedKFold(n_splits = 5 , shuffle = True , random_state = 42)
classifier4 = RandomizedSearchCV ((model_svm) , {
    'C' : range(1 , 50),
    'kernel' : ['rbf' , 'poly' , 'linear' , 'sigmoid'],
    'gamma' : ['scale' , 'auto'],
    'decision_function_shape' : ['ovo' , 'ovr']
} , n_iter = 10 , cv=sf , return_train_score = False)

classifier4.fit(X , Y)
```

# **Leave One Out Cross Validation (LOOCV)**

```python
from sklearn.model_selection import LeaveOneOut , RandomizedSearchCV
loo = LeaveOneOut()
classifier5 = RandomizedSearchCV ((model_svm) , {
    'C' : range(1 , 50),
    'kernel' : ['rbf' , 'poly' , 'linear' , 'sigmoid'],
    'gamma' : ['scale' , 'auto'],
    'decision_function_shape' : ['ovo' , 'ovr']
} , n_iter = 10 , cv=loo , return_train_score = False)

classifier5.fit(X , Y)
```

```python
import optuna
from sklearn.model_selection import cross_val_score
from sklearn.neighbors import KNeighborsClassifier

def objective_knn(trial):
  # suggest_int to specify a range (low, high)
  k = trial.suggest_int('k' , 1 , 10) # Assuming range from 1 to 10 based on original arguments
  weights = trial.suggest_categorical('weights' , ['uniform' , 'distance'])
  p = trial.suggest_int('p' , 1 , 4) # Assuming range from 1 to 4 based on original arguments
  algorithm = trial.suggest_categorical('algorithm' , ['auto' , 'ball_tree' , 'kd_tree' , 'brute'])
  leaf_size = trial.suggest_int('leaf_size' , 10 , 50)

  model = KNeighborsClassifier(n_neighbors = k , weights = weights , p = p , algorithm = algorithm , leaf_size = leaf_size)
  # Perform cross-validation and return the mean score for Optuna to maximize
  scores = cross_val_score(model, X, Y, cv=5, scoring='accuracy')
  return scores.mean()

study_knn = optuna.create_study(direction = 'maximize')
study_knn.optimize(objective_knn , n_trials = 10)

print("Best KNN params:" , study_knn.best_params)
print("Best KNN accuracy:" , study_knn.best_value)
```
