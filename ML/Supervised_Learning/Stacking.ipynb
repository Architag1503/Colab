```python
import numpy as np
import pandas as ps
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
df = sns.load_dataset('iris')
df.head()
```

```python
df.describe()
```

```python
df.info()
```

```python
df.isna().sum()
```

```python
X = df.drop(['species'] , axis=1)
Y = df['species']
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X = scaler.fit_transform(X)
X
```

```python
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import StackingClassifier

X_test , X_train , Y_test , Y_train = train_test_split(X , Y , test_size=0.2 , random_state=42, shuffle=True)
```

```python
base_models = [
    ('svm', SVC(kernel='rbf' , shrinking=True)),
    ('rf', RandomForestClassifier(n_estimators=100 , criterion='gini' , min_samples_split=10 , min_samples_leaf=3)),
    ('dt', DecisionTreeClassifier()),
    ('knn', KNeighborsClassifier(n_neighbors=9 , weights='distance' , leaf_size=30))
]
```

```python
meta_model = LogisticRegression()
```

```python
stack_model = StackingClassifier(estimators=base_models , final_estimator=meta_model , cv=5)
```

```python
stack_model.fit(X_train , Y_train)
```

```python
accuracy = stack_model.score(X_test , Y_test)
print('Stacking Accuracy: ' , accuracy)
```
