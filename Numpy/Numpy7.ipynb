# **Mathematical Geometrical Functions**

```python
#sin , cos , tan , sinh , cosh , tanh FUNCTIONS
#These functions is use to calculate the rrignometruc tangent for all the elements of the array passed as the argument
import numpy as np
import math
arr = np.array([1 , 2 , 3])
print(arr)
res = np.cos(arr)
print(res)
res = np.sin(arr)
print(res)
res = np.tan(arr)
print(res)
```

```python
#Numpy arccos() , arcsin() & arctan() method
#This function is used to calculate the inverse cos , sin & tan of the array elements
import numpy as np
arr = [0 , 0.3 , -1]
print(arr)
x = np.arccos(arr)
print(x)
x = np.arcsin(arr)
print(x)
x = np.arctan(arr)
print(x)
```

```python
#Numpy degree() method
#This function is used to convert the angles from radian to degrees

import numpy as np
import math

arr = [1,2,3]
print(arr)
degval = np.degrees(arr)
print(degval)
```

```python
#Numpy deg2rad() method
#This function is used to convert the angles from degree to radian
import numpy as np
arr = [1,2,3]
print(arr)
degval = np.deg2rad(arr)
print(degval)
```

```python
#Numpy rad2deg() method

import numpy as np
import math

arr = [0 , 1 , 5]
print("Input arr : \n" , arr)

degval = np.rad2deg(arr)
print("\n Degree Value: \n" , degval)
```

```python
#Numpy radians() method
import numpy as np

arr = [0 , 30 , 60 , 90]
print("Input array : \n" , arr)

radval = np.radians(arr)
print("\n Radian value: \n" , radval)
```

```python
#Numpy hypot() method
#This function is used to calculate the hypotenude for the right angled triangle.
#The hypotenuese of the right anfle triangle calculated as:
#h = sqrt(b^2 + p^2)
#where b & p are base & perpendicular of right angle triangle

#Each value of h is copied into the output array

import numpy as np

base = [10 , 2 , 5 , 50]
per = [3 , 10 , 23 , 6]


print("Input base array:" , base)
print("Input perpendicular array:" , per)

hyp = np.hypot(base , per)
print("hypotenuse: " , hyp)
```

```python
#Numpy ceil() , Floor() , round()
#This function returns the ceil value of the input array elements. The floor of a number x is i

#Numpy rint()
#This function is used to round the array elements to the nearest integer

arr = [0.23 , 0.69 , 1.2 , 1.24 , 9.99 , 7.7]
print("Input Array: " , arr)
print("Ceil value: " , np.ceil(arr))
print("Floor value: " , np.floor(arr))
print("Round value: " , np.round(arr))
print("Rint value: " , np.rint(arr))
```

```python
#Numpy fix()
#This function is used to round the array values to the nearest integers towards zero.

import numpy as np

arr = [0.23 , 0.09 , 1.2 , 1.24 , 9.99]

print("Input array: " , arr)

r_arr = np.fix(arr)

print('Ouput array: ' , r_arr)
```

```python
#Numpy trunc()
#The function returns the truncated value of the input array elements
#The truncated value t of input value x is the nearest integer closer to zero then x
#The fractional part of the input number is discarded by this function

import numpy as np
arr = [0.23 , 0.09 , 1.2 , 1.24 , 9.99]

print("Input array: " , arr)

r_arr = np.trunc(arr)

print("Output array: " , r_arr)
```
