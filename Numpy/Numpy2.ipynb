**Stacking: Several arrays can be stacked together along different axes**

np.vstack: To stack arrays along vertical axis <br>
np.hstack: To stack arrays along horizontal axis <br>
np.column_stack: To stack 1-D arrays as columns into 2-D arrays<br>
np.concatenate: To stack arrays on specified axis (axis is passed by argument)


```python
import numpy as np

a = np.array([[1,2],
              [3,4]])

b = np.array([[5,6],
              [7,8]])

#vertical stacking
print(np.vstack((a,b))) #merging the two arrays in vertical manner

#horizontal stacking
print(np.hstack((a,b))) #merging the two arrays in horizontal manner

#stacking columns
c = [10,5]
print(np.column_stack((a,c))) #stacking the columns

#contatenation method
print(np.concatenate((a,b),1)) #stacking along the axis 1 represent horizontal stacking & axis 0 reprsent vertical stacking
```

Splitting: For splitting we have these functions:

np.hsplit: Split array along horizontal axis<br>
np.vsplit: Split array along vertical axes<br>
np.array_split: Split array along specified axis

```python
a = np.array([[1,2,3,4,5,7],
              [8,7,6,5,9,4],
              [9,5,2,3,1,8],
              [3,4,5,6,7,1]])

#horizontal splitting
print(np.hsplit(a,3)) #Basically it split the arry along columns

#vertical splitting
print(np.vsplit(a,2)) #basically split the array along rows
```

```python
c = np.array([[1,2,3,4,5,6,7,8]]);
print(np.split(c,2, axis=1))
```

## ***Broadcasting***
The term braodcasting describes how numoy treates arrays with different shapes during arithmatic operations. Subject to certain constraints, the smaller array is "broadcast" across the larger array so that thay have compatible shapes.

```python
a = np.array([1.0,2.0,3.0])
b = 2.0
print(a*b)

c = [2.0,3.0,4.0]
print(a*c)
```

```python

a = np.array([1.0,2.0,3.0,4.0])
b = np.array([0.0,1.0,2.0])

print(a[: , np.newaxis] + b)

'''
Understanding a[:, np.newaxis]
a[:, np.newaxis]


This means:

: → take all elements of a

np.newaxis → add a new column dimension

Before:
a.shape → (4,)

After:
a[:, np.newaxis].shape → (4, 1)


So a becomes:

[[1.],
 [2.],
 [3.],
 [4.]]


A column vector.

Now look at b
b.shape → (3,)


NumPy treats it like a row vector:

[0. 1. 2.]

Broadcasting during addition

Now NumPy does broadcasting:

a (4×1)	b (3,)	result
rows = 4	columns = 3	output = (4×3)

NumPy expands them to:

[[1.]      +  [0. 1. 2.]
 [2.]
 [3.]
 [4.]]
'''
```

```python
a = np.array([[0.0,0.0,0.0],[10.0,10.0,10.0],[20.0,20.0,20.0],[30.0,30.0,30.0]])
b = np.array([1.0,2.0,3.0])

print(a)
print(b)
print(a + b)
```

# **Data Time Working in Numpy**

```python
#creating a date
today = np.datetime64('2017-02-12')
print("Date is:" , today)
print("Year is:" , np.datetime64(today, 'Y'))
```

```python
#creating array of dates in a month
dates = np.arange('2017-02' , '2017-03' , dtype='datetime64[D]')
print("\nDates of February, 2017:\n" , dates)
print("\nToday is February:", today in dates)
```

```python
#arithmatic operation on dates
dur = np.datetime64('2017-05-22') - np.datetime64('2016-05-22')
print("\nNo. of days:", dur)
print("No. of weeks:", np.timedelta64(dur, 'W'))
```

```python
#sorting dates
a = np.array(['2017-02-12' , '2016-10-13' , '2019-05-22'] , dtype='datetime64')
print("\nDates in sorted order:", np.sort(a))
```
