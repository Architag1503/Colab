```python
#numpy.log() in Python
#The numpy.log() is a mathematical function that is use to calculate the natural loagrothms of x (x belongs to all input array elements)
#It is the inverse of the exponential function as well as element-wise natural logarithm. The natural logarithm log is the reverse of the exponential
#so that log(exp(x)) = x. The logarithm in abse e is the natural number

#Syntax
#np.log(x, /, out=None, *, where=True, casting='same_kind', order='K', dtype=None, subok=True)
```

```python
import numpy as np

a = np.array([2,4,6,3**8])
print(a)
b = np.log(a)
print(b)
c=np.log2(a)
print(c)
d=np.log10(a)
print(d)
```

```python
import numpy as np
import matplotlib.pyplot as plt

arr = [2 , 2.2 , 2.4 , 2.6 , 2.8 , 3]
result1=np.log(arr)
result2=np.log2(arr)
result3=np.log10(arr)
plt.plot(arr , arr , color='blue' , marker="*")
plt.plot(result1 , arr , color='green' , marker="o")
plt.plot(result2 , arr , color='red' , marker="*")
plt.plot(result3 , arr , color='black' , marker="*")
plt.show()
```

```python
#numpy.where in python
#The Numpy module provides a function numpy.where() for selecting elements based on a condition
#It return elements chosen from a or b depending on the condition

#For example , all arguments -> condition, a & b are passed in numpy.where() then it will return elements selected from a & b
#depending on values in bool array yielded by the condition

#If only the condition is provided, this function is a shorthand to the function np.asarray (condition).nonzero().
#Although nonzero should be preferred directly, as it behaves correctly for subclasses

#Syntax:
#numpy.where(condition[ , x, y])
```

```python
import numpy as np
a = np.arange(10)
b = np.where(a < 5 , a , 10*a)
b
```

```python
#Broadcasting x,y and condition
import numpy as np
b = np.where([[True , False], [True , False]] , [[1,2], [3,4]] , [[9,8] , [7,6]])
b
#Position	Condition	Value  taken-From
#(0,0)	   True	     1	   x[0][0]
#(0,1)	   False	   8	   y[0][1]
#(1,0)	   True	     3	   x[1][0]
#(1,1)	   False	   6	   y[1][1]
```

```python
#numpy.argsort() in Python
#The numpy module provides the function argsort() , returns the indices which would sort an array

#The NumPy module provides a function for performing an indirect sort along with the given axis witht the help of algorithm specified
#This function returns an array of indices of the same shape as 'a' , which would sort the array

#Syntax
#numpy.argsort(a , axis=-1 , kind=None , order=None)
```

```python
import numpy as np
a = np.array([456, 11 , 63])
a
b = np.argsort(a)
b
```

```python
import numpy as np
a = np.array([[0,5] , [3,2]])
indices = np.argsort(a , axis=0)
print(indices)
```

```python
import numpy as np
a = np.array([[0,5] , [3,2]])
indices = np.argsort(a , axis=0)
indices
np.take_along_axis(a , indices , axis=0)
```

```python
import numpy as np
a = np.array([[0,5] , [3,2]])
indices = np.argsort(a , axis=1)
print(indices)
```

```python
import numpy as np
a = np.array([[0,5] , [3,2]])
indices = np.argsort(a , axis=1)
indices
np.take_along_axis(a , indices , axis=1)
```

```python
import numpy as np
a = np.array([[0,5] , [3,2]])
indices = np.unravel_index(np.argsort(a , axis=None) , a.shape)
indices
a[indices] # same as np.sort(a , axis=None)
```

```python
import numpy as np
a = np.array([(0,5) , (3,2)] , dtype=[('x' , '<i4') , ('y' , '<i4')])
print(a)
b = np.argsort(a , order=('x' , 'y'))
print(b)
c = np.sort(a , order = ('y' , 'x'))
print(c)
```

```python
#numpy.transpose() in Python
#The numpy.transpose() function is one of the most important function in matrix multiplication
#The function permutes or reserve the dimension of the given array and return the modified array

#The numpy.transpose() function changes the row elemnts into column elements and the column
#elements into row elements. The output of this function is a modified array of the original one.

#Syntax
#numpy.transpose(a , axis=None)
```

```python
import numpy as np
a = np.arange(12).reshape(3,4)
print(a)
b = np.transpose(a)
print(b)
```

```python
import numpy as np
a = np.array([[1,2] , [2,3] , [3,4]])
print(a)
b = np.transpose(a , (1,0))
print(b)
```

```python
import numpy as np
#np.ones(...) → creates an array filled with 1.0
#Shape is 4-D:
a = np.ones((12,32,123,64))
print(a)
#axis 0 → size 12
#axis 1 → size 32
#axis 2 → size 123
#axis 3 → size 64

b = np.transpose(a , (1,3,0,2)).shape
print(b)

#New axis position	Comes from old axis	 Old size
# new axis 0           old axis 1	          32
# new axis 1	         old axis 3	          64
# new axis 2	         old axis 0	          12
# new axis 3	         old axis 2	          123

c = np.transpose(a , (0,3,1,2)).shape
print(c)
#New axis position	 Comes from old axis	Old size
#new axis 0	           old axis 0	          12
#new axis 1	           old axis 3	          64
#new axis 2	           old axis 1	          32
#new axis 3	           old axis 2	          123
```

```python
#numpy.mean() in Python
#The sum of elements , along with axis divided by the number of elements , is known as arithmatic mean.
#The funcyion is used to compute the arithmatic mean along the specified axis

#This function returns the average of the array elements. By default , the average is taken on the flatteed array
#float64 is intermediate as well as return values are used for integer inputs

#Syntax
#numpy.mean(a , axis=None , dtype=None , out=None , keepdims=<no value>)
```

```python
import numpy as np
a = np.array([[1,2], [3,4]])
b = np.mean(a);
print(b)
x = np.array([[5,6],[7,34]])
y = np.mean(x)
print(y)
```

```python
import numpy as np
a = np.array([[1,2], [3,4]])
b = np.mean(a , axis=0)
print(b)
x = np.array([[5,6],[7,34]])
y = np.mean(x , axis=1)
print(y)
```

```python
import numpy as np
a = np.zeros((2 , 512*512) , dtype=np.float32)
a[0 , :] = 23.0
a[1 , :] = 32.0
b = np.mean(a)
print(b)
```
