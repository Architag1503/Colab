Numpy is a general purpose array processing package/libraray. It provides a high performance multidimensional array object and tools for working with these arrays.

It is fundamental package for scientific computing with python. It contain various features including these important ones.

Numpy's main object is the homogeneous multidimensional array.

It is table of all elements all of the same type, indexed by a tuple of positive integers. In Numpy dimensions are called axes. The number of axes is a rank. Numpy array class is class ndarray. It is also known by the alias array.

slicing is a way to extract a subsequence (like a part of a list, tuple, string, or any sequence type) from a sequence. It allows you to select elements by specifying start, stop, and step indices.

sequence[start:stop:step]
Where:

start → index to start the slice (inclusive, default is 0)

stop → index to end the slice (exclusive)

step → how many steps to jump between elements (default is 1)

```python
arr = np.array([[1,2,3,4],
                [4,5,6,7],
                [8,7,6,5]])

temp = arr[0:3:2]
print(arr)
print(temp)
```

```python
arr = np.array([[1,2,3,4],
                [4,5,6,7],
                [8,7,6,5]])

temp = arr[:2 ,:2] #first factor represent end row and second represent end column
print(arr)
print(temp)
```

```python
#Python program to demonstrate sorting in numpy
import numpy as np

a = np.array([[1,4,2],
              [3,4,6],
              [0,-1,5]])

#sorted array
print("Array elements in sorted order:\n" , np.sort(a , axis=None))

#sort array row-wise
print("Row-wise sorted array:\n" , np.sort(a , axis=1))

```

```python
#specify sort algorithm
print("Column wise sort by applying merge-sort:\n" , np.sort(a , axis=0 , kind='mergesort'))

```

```python
#Example to show sorting of structured array
#set alias names for dtypes
dtypes = [('name' , 'S10') , ('grad_year', int) , ('cgpa' , float)]

#Values put in the array
values = [('Hrithik' , 2009, 8.5) , ('Ajay' , 2008  , 9.7) ,
          ('Pankaj' , 2008 , 7.9) , ('Akash' , 2009 , 9.0)]

#Creating array
arr = np.array(values , dtype=dtypes)

print("\nArray sorted by names:\n" , np.sort(arr , order='name'))
print("\nArray sorted by graduation year\n" , np.sort(arr , order='grad_year'))
print("\nArray sorted by cgpa\n" , np.sort(arr , order="cgpa"))
print("\nArray sorted by graduation year then cgpa\n" , np.sort(arr , order=["grad_year" , "cgpa"]))
```
