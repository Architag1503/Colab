```python
import numpy as np
import pandas as pd
```

# **Finding Missing Data**

```python
data = {
    'A' : [1 , 2 , np.nan , 4 , 5],
    'B' : [1 , 2 , 3 , 4 , 5],
    'C' : [1 , 2 , 3 , np.nan , np.nan],
    'D' : [1 , np.nan , np.nan , np.nan , 5]
}

df = pd.DataFrame(data)
df
```

```python
df.isna()
```

```python
df.isna().sum()
```

```python
df.isna().any() # This is use to check is there any null value exist in each column
```

# **Removing Missing Values**

```python
#Removing missing values work on row basis not on column basis
df
```

```python
df.dropna()# the rows that contains null values are get remove
```

```python
df
```

```python
df.dropna(thresh=3)# the rows that contain non-null values less than thresh value are get removed
```

```python
df.dropna(thresh=4)
```

# **Filling the Missing Data**

```python
df
```

```python
#shortcut to learn
#for removing -> df.dropna()
#for filling -> df.fillna()
#for searching -> df.isna()
```

```python
df.fillna(0)
```

```python
values = {'A':0 , 'B':100 , 'C':300 , 'D':400}
df.fillna(value=values)
```

```python
#Concept
#Selection & filling will be done column-wise and removal will be done row-wise
```

```python
df
```

```python
df.fillna(df.mean())
```
