**DataFrame** is a 2-dimensional, labeled data structure used to store and manipulate tabular data (like Excel sheets or SQL tables).

Rows → records / observations<br>
Columns → features / variables<br>
Each column has a name<br>
Rows have index values

```python
import numpy as np
import pandas as pd
```

**Creating a Dataframes**

```python
data = {
    'Name' : ['John' , 'Anna' , 'Peter' , 'Linda'],
    'Age' : [28 , 34 , 29 , 42],
    'Location' : ['New York' , 'Paris' , 'Berlin' , 'London'],
    'Salary' : [65000 , 70000 , 62000 , 85000]
}

df = pd.DataFrame(data)
df
```

```python
data_list = [
    ['John' , 28 , 'New York' , 65000],
    ['Anna' , 34 , 'Paris' , 70000],
    ['Peter' , 29 , 'Berlin' , 62000],
    ['Linda' , 42 , 'London' , 85000],
]
```

```python
df = pd.DataFrame(data_list) #columns name are not come
df
```

```python
df2 = pd.DataFrame(data_list)
columns = ['Name' , 'Age' , 'Location' , 'Salary']
df2 = pd.DataFrame(data_list , columns = columns)#Now names of columns are come
df2
```

**Selection and Indexing of Columns**

```python
df2[['Name' , 'Location']]
```

**Creating a new column**

```python
df2["Designation"] = ["Doctor" , "Engineer" , "CA" , "Politician"]
df2
```

**Removing a Column**

```python
df2.drop(["Location" , "Salary"] , axis = 1 , inplace=True) #To delete from main data we have to use inplace
df2
```

**Removing Rows**

```python
df2.drop(0 , axis = 0 , inplace=True) #To delete from main data we have to use inplace
df2
```

**Selecting Rows**

```python
df2.loc[1] # Since we had remove row 0 above from main memory so we can't access it
```

```python
df2.loc[[1 , 3]]
```

```python
df.iloc[3]
```

**Selecting Subsets of Rows and Columns**

```python
df.loc[[0 , 1]][["Location" , "Salary"]]
```

```python
df2.loc[[2 , 3]][["Name" , "Designation"]]
```

**Conditional Selection**

```python
df2
```

```python
#I only want to see those people whose age is above 30
```

```python
df2[df2["Age"] > 30]
```

```python
#I only want those people whose age is above 30 and their city must be Paris
```

```python
df2[(df2['Age'] > 30) & (df2['Location'] == 'Paris')] # AND operation
```

```python
df2[(df2['Age'] > 30) | (df2['Location'] == 'Paris')] # OR operation
```

```python
df2[(df2['Age'] > 30) ^ (df2['Location'] == 'Paris')] # XOR operation
```
