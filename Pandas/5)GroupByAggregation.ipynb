```python
import numpy as np
import pandas as pd
```

```python
data = {
    'Category' : ['A' , 'B' , 'A' , 'B' , 'A' , 'B' , 'A' , 'B'],
    'Store' : ['S1' , 'S1' , 'S2' , 'S2' , 'S1' , 'S2' , 'S2' , 'S1'],
    'Sales' : [100 , 200 , 150 , 250 , 120 , 180 , 200 , 300],
    'Quantity' : [10 , 15 , 12 , 18 , 8 , 20 , 15 , 25],
    'Date' : pd.date_range('2023-01-01' , periods=8)
}

df = pd.DataFrame(data)
```

```python
df
```

```python
#Group By category & calculate the sum of sales
x = df.groupby('Category')['Sales'].sum()

print(x)
```

```python
#Group by Store and calculate the sum of sales
x = df.groupby('Store')['Sales'].sum()
print(x)
```

```python
#Group by Multiple columns
x = df.groupby(['Category' , 'Store'])

for group_name , group_data in x:
    print(group_data)
    print(group_name)
```

```python
x = df.groupby(['Category' , 'Store'])['Sales'].sum()
print(x)
```

# **Aggregation**

```python
x = df['Sales'].mean()
print(x)
```

```python
x = df['Sales'].median()
print(x)
```

```python
x = df['Sales'].mode()
print(x)
```

```python
x = df['Sales'].std()
print(x)
```

```python
x = df['Sales'].var()
print(x)
```

```python
df['Sales'].agg(['sum' , 'mean' , 'median' , 'min' , 'max', 'std' , 'var'])
```
