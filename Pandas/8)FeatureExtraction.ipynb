```python
#Q1: Make a new column of episode count
#If we see the episodes are written in title in ()

def extraxt_episode(txt):
  check = False
  data = ""
  for i in txt:
    if i == ')':
      return data
    if check == True:
      data += i
    if i == '(':
      check = True


df["episode_count"] = df["Title"].apply(extraxt_episode)
df["episode_count"]
```

```python
#Q:6 Which is the longest running anime

def extract_anime(max_idx , df):
  data = ""

  for i in df.loc[max_idx , "Title"]:
      if i == '(':
         break
      else:
         data += i

  return data

#method 1
month_map = {
    "Jan": 1, "Feb": 2, "Mar": 3, "Apr": 4,
    "May": 5, "Jun": 6, "Jul": 7, "Aug": 8,
    "Sep": 9, "Oct": 10, "Nov": 11, "Dec": 12
}

def calculate_duration(time_txt):
  if not isinstance(time_txt, str):
        return 0

  start , end = time_txt.split(' - ')

  start_month , start_year = start.split()
  end_month , end_year = end.split()

  return (int(end_year) - int(start_year))*12 + (month_map[end_month] - month_map[start_month])

df["duration_month"] = df["time_stamp"].apply(calculate_duration)

x = df[["duration_month" , "Title"]]

x = x.sort_values(by="duration_month" , ascending=False).reset_index(drop = True)

print(extract_anime(0 , x))


#method 2
dates = df["time_stamp"].str.split(" - " , expand=True)

start = pd.to_datetime(dates[0] , format="%b %Y")
end = pd.to_datetime(dates[1] , format="%b %Y")

df["duration_month"] = (end.dt.year - start.dt.year)*12 + (end.dt.month - start.dt.month)

x = df[["duration_month" , "Title"]]
x = x.sort_values(by="duration_month" , ascending=False).reset_index(drop = True)

print(extract_anime(0 , x))
```
