```python
#Q11 : which country in africa region has highest population

#method 1
# Step 1: identify Africa rows
africa_mask = df["region"].str.split().apply(lambda x: "Africa" in x)

# Step 2: max population only within Africa
max_pop_africa = df.loc[africa_mask, "population"].max()

# Step 3: final result
result = df[ africa_mask & (df["population"] == max_pop_africa)]["country"]

print(result)

#method 2
cnt = 0

def extract_africa(txt):
    global cnt
    x = txt.split()

    if "Africa" in x:
        cnt += 1
        return True
    return False

# Step 1: Africa rows
africa_df = df[df["region"].apply(extract_africa)]

# Step 2: Africa max population
max_pop_africa = africa_df["population"].max()

# Step 3: Final answer
result = africa_df[africa_df["population"] == max_pop_africa]["country"]

print(result)
```
