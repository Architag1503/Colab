# **Pandas Series**
A series is a one-dimensional labelled array capable of holding any data type. The axis labels are collectively called the index

# **Creating Series**
There are multiple ways of creating the series

```python
import numpy as np
import pandas as pd
```

```python
labels = ['a' , 'b' , 'c']
my_list = [10 , 20 , 30]
arr = np.array([10 , 20 , 30])
d = {'x':10 , 'y':20 , 'z':30}
```

```python
pd.Series(my_list)
```

```python
pd.Series(my_list , index=labels)
```

```python
pd.Series(arr , index=labels)#only 1D array can used because Series are 1D
```

```python
pd.Series(d)#Dictionary keys are behaving as index or labels in Series
```
