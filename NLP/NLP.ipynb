```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("/content/drive/MyDrive/dataset/Emotion_Dataset/train.txt" , sep=';' , header=None , names=['text' , 'emotion'])
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
unique_emotions = df['emotion'].unique()
emotion_number = {}
i = 0
for emotion in unique_emotions:
  emotion_number[emotion] = i
  i += 1

df['emotion'] = df['emotion'].map(emotion_number)
df
```

```python
# step 1 = Covert into lower case
df['text'] = df['text'].apply(lambda x : x.lower())
```

```python
# step 2 = remove punctuations
import string
def remove_punctuation(text):
  for punctuation in string.punctuation:
    text = text.replace(punctuation , '')
  return text

df['text'] = df['text'].apply(remove_punctuation)
```

```python
# step 3 = remove numbers
def remove_numbers(txt):
  new = ""
  for i in txt:
    if not i.isdigit():
      new += i
  return new

df['text'] = df['text'].apply(remove_numbers)
```

```python
#step 4 = remove URL's & Links
def remove_links(text):
  import re
  url_pattern = re.compile(r'https?://\S+|www\.\S+')
  return url_pattern.sub(r'', text)

df['text'] = df['text'].apply(remove_links)
```

```python
# step 5 = remove emojis
def remove_emojis(text):
  new = ""
  for i in text:
    if i.isascii():
      new += i

  return new

df['text'] = df['text'].apply(remove_emojis)
```

```python
# step 6 = remove stopwords (is , am , the , are , that etc.)
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
nltk.download('stopwords')
nltk.download('punkt')
```

```python
stop_words = set(stopwords.words('english'))
def remove(text):
  words = text.split()
  cleaned = []
  for i in words:
    if i not in stop_words:
      cleaned.append(i)

  return ' '.join(cleaned)

df['text'] = df['text'].apply(remove)
```

```python
# Feature Extraction
# Technique 1 : Bag of Words
from sklearn.feature_extraction.text import CountVectorizer

documents = [
    "I love pizza",
    "Pizza is the best",
    "I love pasta",
    "Pasta is great"
]

vectorizor = CountVectorizer(ngram_range = (1,3))

X = vectorizor.fit_transform(documents)

print("Vocabulary: " , vectorizor.get_feature_names_out())
print("\nBox Matrix: \n" , X.toarray())
```

```python
# Technique 2 : TF-IDF
from sklearn.feature_extraction.text import TfidfVectorizer

documents = [
    "I love pizza",
    "Pizza is the best",
    "I love pasta",
    "Pasta is great"
]

vectorizor = TfidfVectorizer()

X = vectorizor.fit_transform(documents)

print("Vocabulary: " , vectorizor.get_feature_names_out())
print("\nBox Matrix: \n" , X.toarray())
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(df['text'], df['emotion'], test_size=0.20, random_state=42)
```

```python
X_train
```

```python
# Step 3 - Feature Extraction from dataset
from sklearn.feature_extraction.text import TfidfVectorizer , CountVectorizer
```

```python
# back of words
bow_vectorizer = CountVectorizer()
X_train_bow = bow_vectorizer.fit_transform(X_train)
X_test_bow = bow_vectorizer.transform(X_test)
```

```python
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score, classification_report

nb_model = MultinomialNB()
nb_model.fit(X_train_bow, y_train)
pred_nb = nb_model.predict(X_test_bow)

print("Accuracy:", accuracy_score(y_test, pred_nb))
print("\nClassification Report:\n", classification_report(y_test, pred_nb))
```

```python
# tf-idf
tfidf_vectorizer = TfidfVectorizer()
X_train_tfidf = tfidf_vectorizer.fit_transform(X_train)
X_test_tfidf = tfidf_vectorizer.transform(X_test)

nb2_model = MultinomialNB()
nb2_model.fit(X_train_tfidf, y_train)
pred_nb = nb2_model.predict(X_test_tfidf)

print("Accuracy:", accuracy_score(y_test, pred_nb))
print("\nClassification Report:\n", classification_report(y_test, pred_nb))
```

```python
from sklearn.linear_model import LogisticRegression

logistic_model = LogisticRegression(max_iter=1000)
logistic_model.fit(X_train_tfidf, y_train)
pred_logistic = logistic_model.predict(X_test_tfidf)

print("Accuracy:", accuracy_score(y_test, pred_logistic))
```
