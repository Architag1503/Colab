# **Linear Algebra in NumPy**

```python
import numpy as np

A = np.array([[6,1,1],
              [4,-2,5],
              [2,8,7]])

print("\nRank of A:" , np.linalg.matrix_rank(A))
print("\nTrace of A:" , np.trace(A))
print("\nDeterminant of A:\n" , np.linalg.det(A))
print("\nInver of A:\n" , np.linalg.inv(A))
print("\nMatrix A raised to power 3:\n" , np.linalg.matrix_power(A,3))
```

```python
#cofficient
a = np.array([[1,2] , [3,4]])
#constants
b = np.array([8,18])

print("Solution of linear equation:" , np.linalg.solve(a , b))
```

```python
#The Dot product is written using a central dot:
#a . b
#This means the Dot Product of a and b
#We can calculate the Dot Product of two vectors this way
#dot product magnitudes and angle
#a . b = |a| * |b| * cos(θ)

product = np.dot(5 , 4)
print("Dot product of scalar values: " , product)
vector_a = 12
vector_b = 4
product = np.dot(vector_a , vector_b)
print("Dot product of two vectors: " , product)
```

```python
#This function return the dot product of the two vectors. If the argument is complex , then its conjugate id used for dot product
#if the argument id is multi-dimensional array, it is flattened

a = np.array([[1, 2], [3, 4]])
b = np.array([[5, 6], [7, 8]])

print("Dot product of a & b:\n" ,np.dot(a, b))
print("\nVector product of a & b:\n" ,np.vdot(a, b)) #(1*5 + 2*6 + 3*7 + 4*8) = 70

a = np.array([1+2j, 3+4j])
b = np.array([5+6j, 7+8j])

print("\nDot product of a & b:\n" ,np.dot(a, b)) #(1+2j)(5+6j) + (3+4j)(7+8j)
print("\nVector product of a & b:\n" ,np.vdot(a, b)) #conj(a) · b


```

```python
#Inner product of two arrays
a = np.array([1,5,0])
b = np.array([2,4,-1])

print("Inner product of a & b:\n" ,np.inner(a, b)) # 1*2 + 5*4 + 0*(-1) = 22
```

```python
#Compute the Outer product of two vectors
x = np.array(['a' , 'b' , 'c'] , dtype=object)
np.outer(x , [1,2,3])#result[i][j] = x[i] * y[j]
```

```python
#Matrix or vector norms
#This function is able to return one of eight different matrix norms , or one of an infinite
#number of vector norms (described below), depending on the value of ord parameter

from numpy import linalg as LA
a = np.arange(9) - 4
print(a)

b = a.reshape((3, 3))
print(b)
```

```python
print(LA.norm(a))#sqrt((−4)^2+(−3)^2+⋯+4^2) = 7.745..
print(LA.norm(b))#sqrt((−4)^2+(−3)^2+⋯+4^2) = 7.745.. (same as above)
print(LA.norm(b , 'fro')) #same as above
print(LA.norm(a , np.inf)) #∥A∥∞ ​= max(∣ai​∣) -> max(4,3,2,1,0,1,2,3,4)=4
print(LA.norm(b , np.inf)) #∥B∥∞​=max(∑​∣bij​∣) -> Row 1: 4 + 3 + 2 = 9 , Row 2: 1 + 0 + 1 = 2 , Row 3: 2 + 3 + 4 = 9 -> max=9
print(LA.norm(a , -np.inf)) #∥A∥−∞​=min(∣ai​∣) -> min(4,3,2,1,0,1,2,3,4) = 0
print(LA.norm(b , -np.inf)) #∥B∥−∞​=min(j∑​∣bij​∣) -> Row 1: 9 , Row 2: 2 , Row 3: 9 -> min = 2
```

```python
#Compute the conditional number of matrix
#This function is capable of returning the condition number using one if Eight different norms

from numpy import linalg as LA
a = np.array([[1,0,-1] , [0,1,0] , [1,0,1]])
a
```

```python
print(LA.cond(a))#cond(A)=∥A∥⋅∥A^−1∥ -> It measures numerical stability , Larger value ⇒ matrix is ill-conditioned
print(LA.cond(a , 'fro'))#condF​(A)=∥A∥F​⋅∥A^−1∥F​ , ∥A∥F​=sqrt(∑aij^2​)
print(LA.cond(a , np.inf))#∥A∥∞​=max(row sum) , Row 1: 1 + 0 + 1 = 2 , Row 2: 1 , Row3:2  =>  ∥A∥∞​=2
print(LA.cond(a , -np.inf))#∥A∥−∞​=min(row sum)=1
print(LA.cond(a , 1))#1-norm (Max column sum) , Col 1: 2 , Col 2: 1 , Col 3: 2 , ∥A∥1​=2
print(LA.cond(a , -1))#Min column sum norm
print(LA.cond(a , 2))#cond2​(A) = σmin/​σmax​​ = 1.414..
print(LA.cond(a , -2))#cond−2 ​= σmax/​σmin ​​= 0.7071..
```

**numpy.linalg.multi_dot** <br>
Compute the dot product of two or more arrays in single function call , while automatically selecting the fastest evaluation order

```python
from numpy.linalg import multi_dot

#prepare some data
A = np.random.random((10000 , 100))
B = np.random.random((100 , 10000))
C = np.random.random((10000 , 5))
D = np.random.random((5 , 333))

#the actual dot multiplication
#A · B → (10000, 10000)
#(A·B) · C → (10000, 5)
#((A·B)·C) · D → (10000, 333)
print(multi_dot([A , B , C , D]))
```

```python
from numpy import linalg as LA
#creating an array
#function
a = np.array([[1,-2j] , [2j,5]])

print("Array is:\n" , a)

#Calculate the eign values
#Using eigh() function
c , d = LA.eigh(a)

print("Eign value is:\n" , c)
print("Eign value is:\n" , d)
```
