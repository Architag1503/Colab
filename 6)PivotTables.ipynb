```python
import pandas as pd
import numpy as np

data = {
    'Date': pd.date_range('2023-01-01', periods=20),
    'Product': ['A', 'B', 'C', 'D'] * 5,
    'Region': [
        'East', 'West', 'North', 'South',
        'East', 'West', 'North', 'South',
        'East', 'West', 'North', 'South',
        'East', 'West', 'North', 'South',
        'East', 'West', 'North', 'South'
    ],
    'Sales': np.random.randint(100, 1000, size=20),
    'Units': np.random.randint(10, 100, size=20),
    'Rep': [
        'John', 'Marry', 'Bob', 'Alice',
        'John', 'Marry', 'Bob', 'Alice',
        'John', 'Marry', 'Bob', 'Alice',
        'John', 'Marry', 'Bob', 'Alice',
        'John', 'Marry', 'Bob', 'Alice'
    ]
}

df = pd.DataFrame(data)

df['Month'] = df['Date'].dt.month_name()
df['Quarter'] = 'Q' + df['Date'].dt.quarter.astype(str)
df
```

pivot_table is used to summarize and aggregate data in a table format (similar to Excel Pivot Tables).

It allows:

Grouping data

Applying aggregation functions

Reshaping data for analysis

Syntax:<br>
pd.pivot_table(
    data,
    values=None,
    index=None,
    columns=None,
    aggfunc='mean',
    fill_value=None,
    margins=False
)

```python
pd.pivot_table(df , values= "Sales" , index="Region", columns="Product" , aggfunc = "median")
```

```python
pivot2 = pd.pivot_table(df , values=["Sales" , "Units"] , index="Region" , columns="Product")
pivot2
```

# **Cross Tabs**

crosstab is used to compute frequency tables (counts) between categorical variables.

pd.crosstab(
    index,
    columns,
    values=None,
    aggfunc=None,
    margins=False,
    normalize=False
)


```python
pd.crosstab(df['Region'] , df['Product'])
```
