## Model Definition and Training
This section defines, compiles, and trains the Recurrent Neural Network (RNN) and Long Short-Term Memory (LSTM) models for word prediction.

```python
embedding_dim = 50
rnn_units = 128 #hidden layer neurons
```

### Set Embedding and RNN Units Dimensions
Defines the `embedding_dim` (the size of the vector space in which words will be embedded) and `rnn_units` (the number of units in the SimpleRNN and LSTM layers, representing the hidden state size).

```python
rnn_model = Sequential()

rnn_model.add(
    Embedding(input_dim=vocab_size , output_dim=embedding_dim , input_length=max_len)
)

rnn_model.add(SimpleRNN(units=rnn_units))
rnn_model.add(Dense(units=vocab_size , activation='softmax'))
```

### Define SimpleRNN Model
Constructs a sequential RNN model:
-   **Embedding Layer**: Maps integer word indices to dense vectors of fixed size (`embedding_dim`). `input_dim` is `vocab_size` (total unique words), `output_dim` is `embedding_dim`, and `input_length` is `max_len` (length of input sequences). This layer learns to represent words in a meaningful vector space.
-   **SimpleRNN Layer**: A basic recurrent layer with `rnn_units` neurons, capable of processing sequences and retaining information from previous steps.
-   **Dense Layer**: The output layer with `vocab_size` units (one for each possible next word) and `softmax` activation to output probabilities for each word in the vocabulary. The word with the highest probability is the model's prediction.

```python
rnn_model.compile(loss='categorical_crossentropy' , optimizer='adam' , metrics=['accuracy'])
```

### Compile SimpleRNN Model
Compiles the RNN model, preparing it for training:
-   **Loss Function**: `categorical_crossentropy` is used as it's suitable for multi-class classification where targets are one-hot encoded. It measures the error between predicted probabilities and true one-hot labels.
-   **Optimizer**: `adam` is a popular and effective optimization algorithm that adapts the learning rate during training.
-   **Metrics**: `accuracy` is tracked to evaluate the model's performance during training and validation.

```python
rnn_model.summary()
```

### Display SimpleRNN Model Summary

```python
lstm_model = Sequential()
lstm_model.add(
    Embedding(input_dim=vocab_size , output_dim=embedding_dim , input_length=max_len)
)

lstm_model.add(LSTM(units=rnn_units))
lstm_model.add(Dense(units=vocab_size , activation='softmax'))
```

### Define LSTM Model
Constructs a sequential LSTM model, similar to the RNN but using an LSTM layer instead of SimpleRNN:
-   **Embedding Layer**: Same as in the RNN model, mapping word indices to dense vectors.
-   **LSTM Layer**: A type of recurrent neural network that is better at learning long-term dependencies than simple RNNs. It includes 'gates' that regulate the flow of information, making it effective for tasks like language modeling.
-   **Dense Layer**: The output layer with `vocab_size` units and `softmax` activation to predict word probabilities.

```python
lstm_model.compile(loss='categorical_crossentropy' , optimizer='adam' , metrics=['accuracy'])
```

## Prediction and Text Generation
This section contains functions for predicting the next word and generating continuous text using the trained models.

```python
from tensorflow.keras.preprocessing.sequence import pad_sequences
```

### Import `pad_sequences`
Imports `pad_sequences` from `tensorflow.keras.preprocessing.sequence` again, although it was already imported earlier (in cell `fUDc0leGG0nw`). This is needed for preprocessing input text for the `predictor` function to ensure all input sequences have a uniform length.

```python
def predictor(model , tokenizer , text , max_len):
  text = text.lower()
  seq = tokenizer.texts_to_sequences([text])[0]
  seq = pad_sequences([seq] , maxlen=max_len , padding='pre')

  pred = model.predict(seq , verbose=0)
  pre_index = np.argmax(pred)
  return index_to_word[pre_index]
```

### Define `predictor` Function
This function takes a trained model, tokenizer, input text, and maximum sequence length, and predicts the next word:
1.  **Lowercase Text**: Converts the input `text` to lowercase for consistency with training data.
2.  **Tokenize Text**: Converts the cleaned input text into a numerical sequence using the pre-fitted `tokenizer`.
3.  **Pad Sequence**: Pads the numerical sequence to `max_len` using `maxlen` (ensuring uniform input size for the model) with `padding='pre'` (adding zeros to the beginning).
4.  **Predict Next Word**: Uses the `model` to predict the probability distribution over all possible next words. `verbose=0` suppresses output during prediction.
5.  **Get Predicted Word**: `np.argmax(pred)` finds the index of the word with the highest predicted probability. This index is then used to retrieve the actual word string from the `index_to_word` dictionary.

```python
seed_text = "life is the best"
next_word = predictor(lstm_model , tokenizer , seed_text , max_len)
print(next_word)
```

### Test `predictor` Function
Demonstrates the `predictor` function by providing a `seed_text` ("life is the best") and then printing the `next_word` predicted by the `lstm_model`.

```python
def generate_text(model , tokenizer , seed_text , max_len ,  n_words ):
  for _ in range(n_words):
    next_word = predictor(model , tokenizer , seed_text , max_len)
    if next_word == "":
      break
    seed_text += " " + next_word
  return seed_text
```

### Define `generate_text` Function
This function generates a sequence of `n_words` starting from a given `seed_text`:
1.  **Iterative Prediction**: It loops `n_words` times, in each iteration calling the `predictor` function to get the next word based on the current `seed_text`.
2.  **Concatenate Words**: The predicted `next_word` is appended to the `seed_text` to form a growing sequence.
3.  **Break Condition**: If the `predictor` returns an empty string (e.g., if it can't confidently predict a word or hits a special token), the generation process stops.
4.  **Return Generated Text**: The complete `seed_text` containing the original phrase and all generated words is returned.

```python
seed = "the meaning of life"
generate_text = generate_text(lstm_model , tokenizer , seed , max_len , 10)
print(generate_text)
```
