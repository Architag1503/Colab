We use Distributuon plots to visualize the distribution of Quantitative data.

Some plots that allow us to visualize the distribution of a dataset. These plots are:



*   displot / histplot
*   jointplot


*   pairplot
*   rugplot




```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

Some datasets are available in seaborn itself , so we directly load from that.

```python
df = sns.load_dataset('tips')
df
```

```python
df['size'].unique()
```

# **Distplot / histplot**

```python
sns.displot(df["total_bill"] , bins=4)
```

```python
sns.displot(df["total_bill"] , bins=20)

# or sns.histplot(df["total_bill"] , bins=20)
```

```python
sns.histplot(df["total_bill"] , bins=20 , kde=True)
```

```python
plt.subplot(1 , 2 , 1)
sns.histplot(df["total_bill"] , bins=20 , kde=True)
plt.subplot(1 , 2 , 2)
sns.histplot(df["tip"] , bins=20 , kde=True)
```

# **Jointplot**

```python
sns.jointplot(x = 'total_bill' , y = 'tip' , data = df , kind = 'scatter')
```

```python
sns.jointplot(x = 'total_bill' , y = 'tip' , data = df , kind = 'kde')
```

```python
sns.jointplot(x = 'total_bill' , y = 'tip' , data = df , kind = 'hex')
```

```python
sns.jointplot(x = 'total_bill' , y = 'tip' , data = df , kind = 'reg')
```

# **PairPlot**

```python
sns.pairplot(df , hue="sex" , palette="rainbow") # hue basically divide the data category wise on specific parameter
```

```python
sns.pairplot(df , hue="size" , palette="rainbow")
```

# **Rugplot**

```python
sns.rugplot(df)
```

```python
sns.rugplot(df["total_bill"])
```
