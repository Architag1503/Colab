```python
import seaborn as sns
import pandas as pd
import plotly.express as px
import cufflinks as cf
```

```python
tips = sns.load_dataset("tips")
tips
```

```python
px.line(tips, y="total_bill").show()
```

```python
df = tips.groupby('day', observed=True)[['total_bill', 'tip']].mean().reset_index()
df
px.bar(df, x='day', y='tip', color='day')
```

```python
px.scatter(tips, x='total_bill', y='tip', color='day')
```

```python
df = tips[['total_bill' , 'day']]
px.box(df , x='total_bill' , y='day' )
```

```python
fig = px.scatter(
    tips,
    x='total_bill',
    y='tip',
    trendline='ols',
    title='Regression Plot: Tip vs Total Bill'
)

fig.show()
```

```python
px.violin(tips , x='day' , y='tip' , box=True)
```

```python
px.histogram(tips , x='total_bill' , color='day' , nbins=30)
```

```python
px.histogram(tips , x='total_bill' , color='day' , nbins=30 , marginal='box' , histnorm='probability')
```
